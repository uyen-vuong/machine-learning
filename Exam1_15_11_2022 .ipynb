```python
import pandas as pd
import numpy as np
```

```python
source = pd.read_csv('Train_samsung.csv')
display(source)
```

```python
source.isna().sum()
```

```python
df = source.copy()
display(df)
```

```python
from random import choice

labels = ['X1', 'X2', 'X5', 'X10']

for i in labels:

    mask = df[i].isna()

    df.loc[mask,[i]] = df.loc[mask,[i]].applymap(lambda _: choice([0,1]))


mask = df['X3'].isna()
df.loc[mask,['X3']] = df.loc[mask,['X3']].applymap(lambda _: choice(['0','1', '2', '3+']))

df['X8'].fillna(round(df['X8'].mean()), inplace = True)
df['X9'].fillna(round(df['X9'].mean()), inplace = True)

display(df)
```

```python
df1 = df.copy()
df['X1'].replace({'Male':0,'Female':1},inplace = True)
df['X2'].replace({'Yes':1, 'No':0}, inplace = True)
df['X4'].replace({'Graduate':1, 'Not Graduate':0}, inplace = True)
df['X5'].replace({'Yes':1, 'No':0}, inplace = True)
df['Class'].replace({'Y':1, 'N':0}, inplace = True)

dummies = pd.get_dummies(df['X11'])
df = pd.concat([df,dummies],axis=1)
df = df.drop('X11', axis = 1)

dummies1 = pd.get_dummies(df['X3'])
df = pd.concat([df,dummies1],axis=1)
df = df.drop('X3', axis = 1)

c = df['Class']
df.drop('Class',axis = 1, inplace = True)
df = pd.concat([df,c],axis = 1)

display(df)
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

X = df.drop('Class', axis = 1)
Y = df['Class']

_X, _X_, _Y, _Y_ = train_test_split(X, Y, test_size = 0.2, random_state = 147)
clf = make_pipeline(StandardScaler(), RidgeClassifier(random_state = 1)).fit(_X, _Y)
pred = clf.predict(_X_)
print("F1 score: {}".format(f1_score(_Y_, pred, average='weighted')))
```
